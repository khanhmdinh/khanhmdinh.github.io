ADLS Access using Service Principal with Key Vault

```python
kv_name = "fabricstaccess"
kv_app_id_name = 'AppIDSecret'
kv_tenantID_name = 'TenantIDName'
kv_keyname = 'KeyName'
storageAccount = "khanhmdinh"
```

```python
spark.conf.set(f"fs.azure.account.auth.type.{storageAccount}.dfs.core.windows.net", "OAuth")
spark.conf.set(f"fs.azure.account.oauth.provider.type.{storageAccount}.dfs.core.windows.net",
               "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set(f"fs.azure.account.oauth2.client.id.{storageAccount}.dfs.core.windows.net", kv_appid)
spark.conf.set(f"fs.azure.account.oauth2.client.secret.{storageAccount}.dfs.core.windows.net", kv_keyname)
spark.conf.set(f"fs.azure.account.oauth2.client.endpoint.{storageAccount}.dfs.core.windows.net",
               f"https://login.microsoftonline.com/{kv_tenantID}/oauth2/token")
```
